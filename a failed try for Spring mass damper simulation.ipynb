```python
from matplotlib import pyplot as plt
import numpy as np

M_matrix = np.eye(3,3)

k_matrix = np.array([[3, -1, 0], [-1, 2, -1], [0, -1, 3]])

static_disp_vec = np.array([[10], [20], [30]])
initial_disp_vec = 5*(np.random.rand(3, 1))
initial_vel_vec = 5*(np.random.rand(3, 1))

nat_freqs, mod_shapes = np.linalg.eig(np.linalg.inv(np.array(M_matrix)).dot(np.array(k_matrix)))

A_coff = []
B_coff = []
for j in range(3):
    A_r = (mod_shapes[:, j].dot(M_matrix).dot(initial_disp_vec))/(mod_shapes[:, j].dot(M_matrix).dot(mod_shapes[:, j]))
    A_coff.append(A_r[0])
    
    B_r = (mod_shapes[:, j].dot(M_matrix).dot(initial_vel_vec))/((nat_freqs[j])*(mod_shapes[:, j].dot(M_matrix).dot(mod_shapes[:, j])))
    B_coff.append(B_r[0])

time = np.linspace(0, 50, 250)
ith_disp_function = np.zeros((3, len(time)))
for i in range(3):
    for j in range(3):
        ith_disp_function[i,:] += (A_coff[j]*np.cos(nat_freqs[j]*time) + B_coff[j]*np.sin(nat_freqs[j]*time))*mod_shapes[i, j]

    ith_disp_function[i,:] += static_disp_vec[i]
#    subplot(length(eigen_vec(:, 1)), 1, i)
#    plot(time, ith_disp_function(i,:))

 #   this is to plot the static indvusal responces along with the dynamic
 #   systen responce



fig = plt.figure()

    
import time as t           
#   this loop should be manually handeled for system having masses other than 3
for l in range(len(time)):
    
#   put 1 inplace of 0 and uncomment the plot lines in the above for loop to just plot the static indisvual responce    
    if 0:
        break
    for i in range(3):  
        ax1 = plt.subplot2grid((3, 2), (i, 0))
        ax1.plot(time, ith_disp_function[i,:])
    
    ax2 = plt.subplot2grid((3, 2), (0, 1), rowspan = 3)
    
    
    
    
    
    ax2.set(xlim=(0.75, 1.25), ylim=((static_disp_vec[0]-10), (static_disp_vec[2]+10)))
    ax2.plot(np.linspace(0.75, 1.25, 50), np.linspace(static_disp_vec[0], static_disp_vec[0], 50), '--k')
    ax2.plot(np.linspace(0.75, 1.25, 50), np.linspace(static_disp_vec[1], static_disp_vec[1], 50), '--k')
    ax2.plot(np.linspace(0.75, 1.25, 50), np.linspace(static_disp_vec[2], static_disp_vec[2], 50), '--k')
    ax2.plot(1, ith_disp_function[0, l],'sk')
    ax2.plot(1, ith_disp_function[1, l],'sk')
    ax2.plot(1, ith_disp_function[2, l],'sk')
    #ax2.plot(np.linspace(1, 1, len(ith_disp_function[2, l])), ith_disp_function[2, l],'sk')
    
    plt.ylabel('Displacement (m)')
    t.sleep(0.000001)
    plt.cla
        
```
