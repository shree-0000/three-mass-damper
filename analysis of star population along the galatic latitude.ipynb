explain the below plot

After ploting the stars in plane 0 and the plane 8 in the (u-g) vs (g-r) space, we can clearly identify which are population I stars i.e the stars in plane 0 (blue colored) and which are population II stars i.e the stars in plane 8 (red colored). 
The pop I stars occupies the upper half graph space and don't follow any clear linear releation. Where as the pop II starslie in the lower half graph space but more importantly they follow a certain pattern. The pop II stars fall along two intersecting lines, hinting the existance of two subpopulations in the pop II stars. Indead, as we would see in the latter plots the there is a clear distenction between the stars along the line 2 and line 3.

```python
#plot u-g vs g-r
plt.figure(figsize=(6,6))
plt.xlabel('u - g')
plt.ylabel('g - r')
plt.scatter(stars0['u-g'], stars0['g-r'], c = 'b', s = 1, label = 'Plane 0')
plt.scatter(stars8['u-g'], stars8['g-r'], c = 'r', s = 1, label = 'Plane 8')

plt.plot([-3, 6], [1.65, 1.65], 'k--')
plt.annotate('line1', [-2.9, 1.7])

plt.plot([-3, 6], [1.38, 1.38], 'g--')
plt.annotate('line2', [-2.9, 1.2])

plt.plot([-1.8, 6], [-1, 2.7], 'c--')
plt.annotate('line3', [-1.4, -0.95])

plt.xlim(-3, 6)
plt.ylim(-1, 5)
plt.legend()

plt.tight_layout()
```
